##7.Test the following hypotheses:

**a.Older individuals may not possess the same level of technological proficiency and may, therefore, lean toward traditional in-store shopping preferences.**

```python
df.info()
```

H0=there is no relation between age and online or offline shopping

Ha=Older individuals may not possess the same level of technological proficiency and may, therefore, lean toward traditional in-store shopping preferences.

```python
from scipy.stats import ttest_ind

df['total_purchases']=df['numstorepurchases']+df['numwebpurchases']+df['numcatalogpurchases']
df_nonzero=df[df['total_purchases']>0].copy()

df['store_share']=df['numstorepurchases']/df['total_purchases']
median_age=df_nonzero['age'].median()
significance_level=0.05

younger=df_nonzero[df_nonzero['age']<median_age]['store_share']
older=df_nonzero[df_nonzero['age']>median_age]['store_share']

t_stat,p_value=ttest_ind(older,younger, equal_var=False)
print(f"t-statistic: {t_stat}")
print(f"p-value: {p_value}")
print("\n")
if(p_value<significance_level):
  print(f"P value is {p_value} hence Reject null hypothesis")
  print("Older individuals may not possess the same level of technological proficiency and may, therefore, lean toward traditional in-store shopping preferences.")
else:
  print(f"p value is {p_value} fail to reject null hypothesis")
  print("there is no relation between age and online or offline shopping")
```

```python
fig,axes=plt.subplots(1,2,figsize=(10,4))
sns.histplot(older,alpha=0.5,label='older',edgecolor='black',ax=axes[0])
sns.histplot(younger,alpha=0.5,label='younger',edgecolor='black',ax=axes[1])
axes[0].set_title("Older")
axes[1].set_title("Younger")
plt.show()
```

```python
sns.scatterplot(data=df_nonzero,x='age',y='store_share',alpha=0.4)
sns.regplot(data=df_nonzero,x='age',y='store_share',scatter=False,color='red')
plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(12,5))
sns.kdeplot(younger,fill=True,color='orange',alpha=0.5,label='younger',ax=axes[0])
sns.kdeplot(older,fill=True,color='blue',alpha=0.3,label='older',ax=axes[1])
axes[0].set_title("Younger")
axes[1].set_title("Older")

axes[0].axvline(younger.mean(),color='red',linestyle='--',linewidth=2,label=f"younger mean: {younger.mean():.2f}")
axes[1].axvline(older.mean(),color='black',linestyle='--',linewidth=2,label=f"older mean: {older.mean():.2f}")

axes[0].legend()
axes[1].legend()

combined_mean=(younger.median()+older.median())/2
axes[0].axvline(combined_mean-t_stat*younger.std()/len(younger)**0.5,linestyle="-",linewidth=2,label='p_value')
axes[1].axvline(combined_mean+t_stat*older.std()/len(younger)**0.5,linestyle="-",linewidth=2)

plt.show()
```

H0=online shopping and children have no relationship

Ha=Customers with children likely experience time constraints, making online shopping a more convenient option.

```python
df[['total no of children','numwebpurchases']]
```

```python
df_no_children=df[df['total no of children']>0]['numwebpurchases']
df_with_children=df[df['total no of children']==0]['numwebpurchases']


ttest,p_value=ttest_ind(df_no_children,df_with_children,equal_var=False)
print(f"t-statistic: {ttest}")
print(f"p-value: {p_value}")
print("\n")

significance_level=0.05

if(p_value<significance_level):
  print(f"P value is {p_value} hence Reject null hypothesis")
  print("Customers with children likely experience time constraints, making online shopping a more convenient option.")
else:
  print(f"p value is {p_value} fail to reject null hypothesis")
  print("online shopping and children have no relationship")
```

**c.Sales at physical stores may face the risk of cannibalization by alternative distribution channels.**

H0=Sales at physical stores has not disturbance from alternative distribution channels

Ha=Sales at physical stores may face the risk of cannibalization by alternative distribution channels.

```python
# Correlation analysis
store_correlations = df[['numstorepurchases', 'numwebpurchases', 'numcatalogpurchases']].corr()
print("Correlation with Store Purchases:")
print(store_correlations['numstorepurchases'].sort_values())

# Regression analysis
import statsmodels.api as sm

X = df[['numwebpurchases', 'numcatalogpurchases']]
X = sm.add_constant(X)
y = df['numstorepurchases']

model = sm.OLS(y, X).fit()
print("\nRegression Results:")
print(model.summary())

# Check for significance
print("\nHypothesis Test Results:")
if model.pvalues['numwebpurchases'] < 0.05 and model.params['numwebpurchases'] < 0:
    print("âœ“ EVIDENCE of web cannibalization (p = {:.4f})".format(model.pvalues['numwebpurchases']))
else:
    print("Ã— NO evidence of web cannibalization")

if model.pvalues['numcatalogpurchases'] < 0.05 and model.params['numcatalogpurchases'] < 0:
    print("âœ“ EVIDENCE of catalog cannibalization (p = {:.4f})".format(model.pvalues['numcatalogpurchases']))
else:
    print("Ã— NO evidence of catalog cannibalization")
```

**Customer Lifetime Value (CLV)**

Estimate how much a customer is worth in the long run

```python
df['Response_Rate']=df['acceptedcmp1']+df['acceptedcmp2']+df['acceptedcmp3']+df['acceptedcmp4']+df['acceptedcmp5']+df['response']
df['Campaigns_sent']=6
df['Total_spending']=df['mntwines']+df['mntfruits']+df['mntmeatproducts']+df['mntfishproducts']+df['mntsweetproducts']+df['mntgoldprods']
```

```python
df['CLV']=df['Total_spending']*df['Response_Rate']/df['Campaigns_sent']
```

```python
sns.set_style('darkgrid')
sns.lineplot(data=df,x='Response_Rate',y='CLV')
```

```python
sns.set_style('darkgrid')

fig,axes=plt.subplots(2,2,figsize=(12,8))
axes=axes.flatten()
sns.lineplot(data=df,x='kidhome',y='CLV',ax=axes[0])
sns.lineplot(data=df,x='Response_Rate',y='CLV',ax=axes[1])
sns.lineplot(data=df,x='teenhome',y='CLV',ax=axes[2])
sns.lineplot(data=df,x='age',y='CLV',ax=axes[3])


plt.tight_layout()
plt.show()
```
